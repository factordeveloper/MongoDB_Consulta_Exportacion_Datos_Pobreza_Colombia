```python
# Importamos pymongo
from pymongo import MongoClient

# Conexión a la base de datos
client = MongoClient('mongodb://localhost:27017/')  # Cambia la URL si es necesario
db = client['nombre_de_tu_bd']  # Cambia el nombre de tu base de datos
collection = db['nombre_de_tu_coleccion']  # Cambia el nombre de tu colección

# Obtener los valores únicos del campo "PERIODO"
def obtener_periodos_unicos():
    periodos = collection.distinct("PERIODO")
    return set(periodos)  # Convertimos a un conjunto para el formato {20121, 20122, ...}

# Llamamos a la función y mostramos el resultado
valores_unicos = obtener_periodos_unicos()
print(valores_unicos)
```
